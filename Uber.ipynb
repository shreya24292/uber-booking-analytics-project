###Data Cleaning

*remove quotes from ID*

```python
df['Booking ID'] = df['Booking ID'].str.replace('"','')
df['Customer ID'] = df['Customer ID'].str.replace('"','')
```

*convert date*

```python
df['Date'] = pd.to_datetime(df['Date'])
```

*create Date-Time Column*

```python
df['DateTime'] = pd.to_datetime(
    df['Date'].astype(str) + ' ' + df['Time']
)
```

###Handle missing values

*ratings*

```python
df['Driver Ratings'] = df['Driver Ratings']
df['Customer Rating'] = df['Customer Rating']
```

*Booking Value & Distance*

```python
df['Booking Value'] = df['Booking Value'].fillna(0)
df['Ride Distance'] = df['Ride Distance'].fillna(0)
```

*Check Duplicates*

```python
df.duplicated().sum()
```

###Feature Engineering

*Hour of day*

```python
df['Hour'] = df['DateTime'].dt.hour
```

*Day name*

```python
df['Day'] = df['DateTime'].dt.day_name()
```

*Month*

```python
df['Month'] = df['DateTime'].dt.month_name()
```

*Weekend Flag*

```python
df['Weekend'] = df['Day'].isin(
    ['Saturday','Sunday']
)
```
